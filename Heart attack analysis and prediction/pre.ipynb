```python
%matplotlib inline
import plotly.express as px
import plotly.io as pio

pio.renderers.default = 'iframe_connected'  # Using iframe for connected Jupyter kernel

# Plot code
fig=px.histogram(df, 
                 x="output",
                 color="sex",
                 hover_data=df.columns,
                 title="Distribution of Heart Diseases",
                 barmode="group")
fig.show()

```
